```python
# Function to scrape data from tweets
def extract_data(path):
    list_data=[]
    l1=os.listdir(path)
    for i in l1:
        filename=path+'/'+i
        fp=open(filename,'r',encoding='utf-8', errors='replace')
        data=fp.readline()
        list_data.append(data)
    return list_data
```

```python
# Train and test positive and negative tweet data scraping
list_data_train_positive=extract_data('tweet/train/positive')
list_data_train_negative=extract_data('tweet/train/negative')
list_data_test_negative=extract_data('tweet/test/negative')
list_data_test_positive=extract_data('tweet/test/positive')
```

```python
# Function to stem the data
def stemmer(data,flag):
    list_stemmed_words=[]
    if flag:
        tokens=nltk.word_tokenize(data)
        for i in tokens:
            temp=stemmer1.stem(i)
            list_stemmed_words.append(temp)
        data1=' '.join(list_stemmed_words)
    else:
        data1=data
    return data1
```

```python
# Function to preprocess data by removing HTML tags, emojis, punctuations etc. Lower casing and calling stem function.
def preprocessing(list_data_train,flag):
    list_data_train_process=[]
    for i in list_data_train:
        temp=emoji.demojize(i)
        soup = BeautifulSoup(temp, "html.parser")
        temp = soup.get_text()
        temp=re.sub(r'[^\w\s]','',temp)
        temp=re.sub(r'\n+','',temp)
        temp=temp.lower()
        if flag==1:
            flag_var_stemmer=True
        else:
            flag_var_stemmer=False
        temp=stemmer(temp,flag_var_stemmer)
        list_data_train_process.append(temp)
    return list_data_train_process
```

```python
# Creating lists of processed data for regular and stem data for both +ve and -ve
list_data_train_positive_process=preprocessing(list_data_train_positive,0)
list_data_train_negative_process=preprocessing(list_data_train_negative,0)
list_data_train_positive_process_stem=preprocessing(list_data_train_positive,1)
list_data_train_negative_process_stem=preprocessing(list_data_train_negative,1)
list_data_test_positive_process=preprocessing(list_data_test_positive,0)
list_data_test_negative_process=preprocessing(list_data_test_negative,0)
list_data_test_positive_process_stem=preprocessing(list_data_test_positive,1)
list_data_test_negative_process_stem=preprocessing(list_data_test_negative,1)
print(list_data_test_positive_process)
```

```python
# Function to create dictionary of unique words in both +ve and -ve tweets
def set1(list_data_train_positive,list_data_train_negative):
    positive_words=[]
    for i in list_data_train_positive:
        word_list_pos=nltk.word_tokenize(i)
        for j in word_list_pos:
            if j not in positive_words:
                positive_words.append(j)
    negative_words=[]
    for i in list_data_train_negative:
        word_list_neg=nltk.word_tokenize(i)
        for j in word_list_neg:
            if j not in negative_words:
                negative_words.append(j)
                
    dict1=dict()
    dict1[0]=dict()
    dict1[1]=dict()
    for i in positive_words:
        dict1[1][i]='1'
    for i in negative_words:
        dict1[0][i]='0'
    return dict1
```

```python
# Creating unified dictionary and unified tweet list by concating both the sets

dict1=set1(list_data_train_positive_process,list_data_train_negative_process)
v_combined_classes=set(dict1[1].keys()).union(set(dict1[0].keys()))
train_tweet_list=list_data_train_positive_process+list_data_train_negative_process
test_tweet_list=list_data_test_positive_process+list_data_test_negative_process
print(v_combined_classes)


dict1_stem=set1(list_data_train_positive_process_stem,list_data_train_negative_process_stem)
v_combined_classes_stem=set(dict1_stem[1].keys()).union(set(dict1_stem[0].keys()))
train_tweet_list_stem=list_data_train_positive_process_stem+list_data_train_negative_process_stem
test_tweet_list_stem=list_data_test_positive_process_stem+list_data_test_negative_process_stem
print(v_combined_classes_stem)
```

```python
#list of unique words of both classes combined for both stem and non-stem data
list_v1=list(v_combined_classes)
list_v1_stem=list(v_combined_classes_stem)
def new_vocab_dict(list_v,dict3):
    for i in list_v:
        dict3[i] = 0
    return dict3
```

```python
#creating vocabulary of words which occured maximum times in any of the tweet
def max_count_vocab(list1,list_v):
    max_count={}
    max_count = new_vocab_dict(list_v, max_count)
    for j in list1:
        tokens=nltk.word_tokenize(j)
        word_count=defaultdict(int)
        for token in tokens:
            word_count[token]=word_count[token]+1
        for k,v in word_count.items():
            if k in max_count:
                max_count[k]=max(word_count[k],max_count[k])
    return max_count
```

```python
#creating a data frame of feature vector that contains columns as words from corpus and rows as tweets.
def freq_matrix(list_v,train_tweet_list):
    freq_2d=pd.DataFrame(columns=list_v)
    max_count3 = max_count_vocab(train_tweet_list,list_v)
    for i in train_tweet_list:
        words = nltk.word_tokenize(i)
        dict3 = {}
        dict3 = new_vocab_dict(list_v, dict3)
        for word in words:
            if word in list_v:
                dict3[word] = dict3[word] + 1
        
        for k,v in dict3.items():
            if max_count3[k]!=0:
                v = v/max_count3[k]

        new_row = pd.Series(dict3)
        freq_2d = pd.concat([freq_2d, new_row.to_frame().T], ignore_index = True)
    
    return freq_2d
```

```python
# Logistic funtion to retrieve y prediciton probabilities.
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# Gradient funtion to minimize weights for logistic regression
def gradient(X, y_true, y_pred):
    return np.dot(X.T, (y_pred - y_true)) / y_true.size
```

```python
# Loss function to retrieve loss for each iteration
def cross_entropy(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Training Function to fit the data and retrieve optimal weights and loss of each iteration
def logisticRegression(X,y_true,iterations,lr):
    m, n = X.shape
    weights = np.zeros(n)
    loss1=[]
    for i in range(iterations):
            dot_product = np.dot(X,weights)
            y_pred = sigmoid(dot_product)
            grad = gradient(X,y_true,y_pred)
            weights = weights-(lr*grad)
            loss=cross_entropy(y_true,y_pred)
            loss1.append(loss)
    return weights
```

```python
# Calling logistic regression function to get optimal weights for stem and non-stem data 
w = logisticRegression(np.array(train_df['vectors'].tolist()),train_df['label'],iterations=1000,lr=0.01)
w

w_stem = logisticRegression(np.array(train_df_stem['vectors'].tolist()),train_df_stem['label'],iterations=1000,lr=0.01)
w_stem
```

```python
# Transform the testing data and retrieving the predicted classes based on threshold value for test data
def predict(Xtest,weights,threshold):
    dot_product = np.dot(Xtest, weights)
    probabilities = sigmoid(dot_product)
    predictions=[]
    for p in probabilities:
        if p >=threshold:
            predictions.append(1)
        else:
            predictions.append(0)
    return predictions
```

```python
# calling functions to get predicted classed for stem and non-stem test data
y_pred = predict(np.array(test_df['vectors'].tolist()),weights=w,threshold=0.5)
y_pred_stem = predict(np.array(test_df_stem['vectors'].tolist()),weights=w_stem,threshold=0.5)
```

```python
# Function to evaluate and retrieve true pos, true neg, false pos and false neg values
def conf_matrix(y_pred_val,y_actual):
    tp=tn=fp=fn=0
    for i in range(0,len(y_pred_val)):
        if y_actual[i]==1 and y_pred_val[i]==1:
            tp+=1
        elif y_actual[i]==0 and y_pred_val[i]==0:
            tn+=1
        elif y_actual[i]==0 and y_pred_val[i]==1:
            fp+=1
        elif y_actual[i]==1 and y_pred_val[i]==0:
            fn+=1
    return tp,tn,fp,fn        
```

```python
# Function to scrape data from tweets
def export_data(path):
    list_data = []
    l1 = os.listdir(path)
    for l in l1:
        fname = path+'/'+l
        f = open(fname,encoding='utf-8',errors='replace')
        data = f.readline()
        list_data.append(data)
    return list_data   
```

```python
# Function to stem the data
stemmer = SnowballStemmer(language='english')
def stemmerMethod(data,flag):
    list_of_stemmed_words = []
    tokens = word_tokenize(data)
   
    for t in tokens:
        if t[0].isupper() and not t.isupper():
            t=t.lower()
        if flag:
            list_of_stemmed_words.append(stemmer.stem(t))
        else:
            list_of_stemmed_words.append(t)
    return list_of_stemmed_words
```

```python
# Function to preprocess data by removing HTML tags, emojis, punctuations etc. Lower casing and calling stem function.
def preProcess(text,stemmerFlag):
    list_preprocessed = []
    
    for t in text:
        t = emoji.demojize(t)
        soup = BeautifulSoup(t,"html.parser")
        t = soup.get_text()
        t = re.sub(r'[^\w\s]','',t)
        t = re.sub(r'\n+','',t)
        t = stemmerMethod(t,stemmerFlag)
        stop_words = list(set(stopwords.words('english')))
        if t not in stop_words:
            list_preprocessed.append(t)
    return list_preprocessed 
    
    
```

```python
# Function to create term frequency of each term present in the document.
#To normalise the value in between 0 and 1, the value is divided by the length of the document
#The result will be the 2d list which contains words(total vocabulary) as columns and documents as rows, if the word is not present 
#in document it will be considered as zero.
def cal_tf(text):
    tf_arr =[]
    for d in text:
        word_tf = defaultdict(float)
        total_len = len(d)
        for t in d:
            word_tf[t]+=1
        for w in word_tf:
            word_tf[w] = word_tf[w]/total_len
        tf_arr.append(word_tf)
    return tf_arr        
```

```python
# Creating lists of processed data for regular and stem data for both +ve and -ve
list_data_train_positive_process=preProcess(list_train_positive,0)
list_data_train_positive_process_stem=preProcess(list_train_positive,1)
list_data_train_negative_process=preProcess(list_train_negative,0)
list_data_train_negative_process_stem=preProcess(list_train_negative,1)
list_data_test_positive_process=preProcess(list_test_positive,0)
list_data_test_positive_process_stem=preProcess(list_test_positive,1)
list_data_test_negative_process=preProcess(list_test_negative,0)
list_data_test_negative_process_stem=preProcess(list_test_negative,1)
```

```python
# Function to create inverse document frequency of each document.
# document frequency of a word is the number of documents in which the word has been occured, it is obtained by dividing the count
# of document in which the word is present by total number of documents
#Inverse document frequency is obtained by taking reciprocal of DF.
# The Result is normalised by taking the logarithm.
def cal_idf(text):
    n= len(text)
    idf_dict = defaultdict(float)
    unique_tokens = set([token for d in text for token in d])
    
    for t in unique_tokens:
        sum_doc =0
        for d in text:
            if t in d:
                sum_doc+=1
        
        idf_dict[t] = math.log(n/sum_doc)
    return idf_dict
        
```

```python
#Function to Create TF-IDF of each term in a document
#It is obtained by taking product of TF of each term of a document and DF of that term
def cal_tfidf(text,max_features):
    tfidf_text = []
    tf = cal_tf(text)
    idf = cal_idf(text)
    for doc_tf in tf:
        tfidf_vector = {}
        for k,v in doc_tf.items():
            tfidf_vector[k] = v*idf[k]
        
        sorted_tfidf = sorted(tfidf_vector.items(), key=lambda item: item[1], reverse=True)
        tfidf_vector = dict(sorted_tfidf[:max_features])
        
        tfidf_vector_final = np.zeros(max_features)

        for i, (k, v) in enumerate(sorted_tfidf[:max_features]):
            tfidf_vector_final[i] = v
        
        tfidf_text.append(tfidf_vector_final)
    return tfidf_text
```

```python
# Sigmoid funtion which will be used as non-linear activation function.
def sigmoid(x):   
    return 1 / (1 + np.exp(-x))
```

```python
# partial derivative of activation function which used while backpropogating and updating weights
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Mean Square error is used as loss function
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# class for defining neural network which contains:
# constructor which provides no of neurons for input, hidden and output layer, learning rate, initial weight and bias initialisation
# forward pass function to calculate output in each layer using random weight
# backward function to back-propogate to each layer from output to input and update weights of each layer using gradient.
# in train function we set no of epochs and at each epoch we update weights, calculate loss using training data
# after weight optimisation we give testing data to predict function for classification
class myNeuralNetwork():
    def __init__(self,inputSize,hiddenSize,outputSize,lr):
        self.inputSize = inputSize
        self.outputSize = outputSize
        self.hiddenSize = hiddenSize
        self.learningrate = lr
        
        self.w1 = np.random.rand(inputSize,hiddenSize)*0.001
        
        self.w2 = np.random.rand(hiddenSize,outputSize)*0.001
        
        self.bias_h = np.zeros(hiddenSize)
        self.bias_o = np.zeros(outputSize)
    
        
    def forwardPass(self,inputs):
        inputs = inputs
        self.inputHidden = np.dot(inputs,self.w1)+self.bias_h
        self.outputHidden = sigmoid(self.inputHidden)
        self.inputOutput = np.dot(self.outputHidden,self.w2)+self.bias_o
        self.output = sigmoid(self.inputOutput)
        return self.output
    
    def backward(self, X, y_true, y_pred):        
        outputError = y_true - y_pred
        outputDelta = outputError * sigmoid_derivative(y_pred)
        
        hiddenError = outputDelta.dot(self.w2.T)
        hiddenDelta = hiddenError * sigmoid_derivative(self.outputHidden)
        
        self.w2 += self.outputHidden.T.dot(outputDelta) * self.learningrate
        self.bias_o+= np.sum(outputDelta, axis=0) * self.learningrate
        
        self.w1 += X.T.dot(hiddenDelta) * self.learningrate
        self.bias_h += np.sum(hiddenDelta, axis=0) * self.learningrate
   
    def train(self, X, y,epochs=1000):
        for epoch in range(epochs):
            y_pred = self.forwardPass(X)
            self.backward(X, y, y_pred)
            
            loss = mse_loss(y, y_pred)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss: {loss}')
                
    def predict(self, X):
        y_pred = self.forwardPass(X)
        predictions = (y_pred > 0.335).astype(int)
        return predictions
```

```python
# Function to produce evaluation matrix with accuracy, precision, recall and F1
def evaluation(tp, fp, fn, tn):
    total = tp+fp+fn+tn
    prc = tp/(tp+fp)
    rc = tp/(tp+fn)
    print("Precision: ", prc)
    print("Recall: ", rc)
    print("F1: ", (2*prc*rc)/(prc+rc))
    print("Accuracy: ", (tp+tn)/total)
    return prc, rc, ((2*prc*rc)/(prc+rc)), ((tp+tn)/total)
```
